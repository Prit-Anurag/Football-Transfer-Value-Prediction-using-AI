```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv("twitt_text.csv")
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
sentiments = []

for index, row in df.iterrows():
    text = row['tweet_text']
    
    score = analyzer.polarity_scores(text)
    compound = score['compound']
    
    # Decide sentiment
    if compound >= 0.05:
        label = "Positive"
    elif compound <= -0.05:
        label = "Negative"
    else:
        label = "Neutral"
    
    sentiments.append(label)
    
    # Print line-by-line result
    print(f"{index+1}. {text} --> {label} ({compound})")
```

```python
df['sentiment'] = sentiments
```

```python
df.to_csv("player_sentiment_output.csv", index=False)
```
